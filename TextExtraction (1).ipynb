## Introduction
In this small project, I will use pytesseract to extract words from a picture.

```python
import pytesseract
import cv2
import matplotlib.pyplot as plt
```

```python
pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files (x86)\\Tesseract-OCR\\tesseract.exe'
```

## Import the image and change it from BGR to RGB color

```python
img = cv2.imread('C://Users//uzair//Pictures/kitchen.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(25,10))
plt.imshow(img)
plt.show()
```

```python
boxes = pytesseract.image_to_data(img)
for x,b in enumerate(boxes.splitlines()):
    if x !=0:
        b = b.split()
        if len(b) ==12:
            x,y,w,h = int(b[6]), int(b[7]), int(b[8]), int(b[9])
            cv2.rectangle(img, (x, y), (w + x,y + h), color=(0, 0, 255), thickness=1)
            cv2.putText(img, b[11], (x,y), cv2.FONT_HERSHEY_PLAIN, 1, (255, 0, 0), 2)
plt.figure(figsize=(25,10))
plt.imshow(img)
plt.show()
```
